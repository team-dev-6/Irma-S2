```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [384, 450, 384]
x = ['Simba', 'Lays', 'Doritos']
# Plot list y
plt.plot(x,y, 'g') #green 
plt.plot(y, 'ro')  #Red dots
# Plot axes labels and show the plot
plt.xlabel('Chips')
plt.ylabel('Profit')
plt.title('Profit of Chips')
plt.show()
```

```python
y = [39, 30, 40]
x = ['Pepper Steak', 'Chicken', 'Steak & Kidney']
# Plot list y
plt.plot(x,y, 'r') #Red
plt.plot(y, 'b*--') #blue, marks
# Plot axes labels and show the plot
plt.xlabel('Pies')
plt.ylabel('Sold for the week')
plt.title('Pies sold for the week')
plt.grid()
```

```python
# Data values for creating a histogram
y = [5, 5, 5]
# Creating a histogram
plt.hist(y, bins= [1, 2, 3, 4, 5])
plt.xlabel('Bins') #Bins
plt.ylabel('Frequency') #Frequency
plt.title('Cost prices of chocolates')
plt.show()
```

```python
# Data values for creating a histogram
y = [1710, 560, 1000]
# Creating a histogram
plt.hist(y, bins= [300, 600, 900, 1200, 1500, 1800],orientation='horizontal', histtype='step', color='orange')
#Swap X & Y Axes around to for horizontal
plt.xlabel('Cupcakes') #Bins
plt.ylabel('Profit') #Frequency
plt.title('Profits for Cupcakes for the week')
plt.show()
```

```python
import random
y = [16, 16, 10]
x = ['Spinach', 'Cabbage', 'Tomato']
# Creating a scatter plot
color = np.random.rand(3)
size = np.random.randint(50, 100, 10)
plt.scatter(x, y,c=color)
# Labeling the plot and showing it
plt.xlabel('Veg')
plt.ylabel('Prices')
plt.title('Price For Veggies')
plt.show()
```
